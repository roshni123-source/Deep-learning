```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from keras.models import Sequential
from keras.layers import *
```

load train & test data

```python
train_df=pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\vs code\data\sale_training.csv')
train_df.head()
```

```python
test_df=pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\vs code\data\sales_testing.csv')
test_df.head()
```

use minmax feature scailing to scale the variable between 0 & 1

```python
scaler = MinMaxScaler(feature_range=(0, 1))
training_scalar=scaler.fit_transform(train_df)
testing_scalar=scaler.fit_transform(test_df)
```

create dataframe from scaled data

```python
scaled_training_df = pd.DataFrame(training_scalar, columns=train_df.columns.values)
scaled_testing_df = pd.DataFrame(testing_scalar, columns=test_df.columns .values)
scaled_testing_df
```

save scaled data into csv form 

```python
scaled_training_df.to_csv("sale_train_scaled.csv", index=False)
scaled_testing_df.to_csv("sale_test_scaled.csv", index=False)
```

load scaled data

```python
train=pd.read_csv('sale_train_scaled.csv')
test=pd.read_csv('sale_test_scaled.csv')
train.head()
```

split train data into x & y

```python
x=train.drop('total_earnings',axis=1)
y=train['total_earnings']
```

create neural network

```python
# Define the model
model = Sequential()
model.add(Dense(50, input_dim=9, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(1, activation='linear'))
model.compile(loss="mean_squared_error", optimizer="adam")
```

```python
train.shape
```

fit the model

```python
model.fit(
    x,
    y,
    epochs=50,
    shuffle=True,
    verbose=2
)
```

split test data into x & y

```python
x_test=test.drop('total_earnings',axis=1)
y_test=test['total_earnings']
```

predict the model

```python
y_pred= model.evaluate(x_test, y_test, verbose=0)
y_pred
```
